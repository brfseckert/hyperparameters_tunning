## Hyperparameters definition

```
def get_hyperparameter():

  bs = random.choice([4, 8, 16, 32, 64, 128, 256])
  init = random.choice(['zeros', 'uniform', 'normal','he_normal', 'lecun_uniform', 'glorot_uniform'])
  lr= round(random.uniform(0.00095, 0.001), 6)
  lr_input = str(lr)
  opt = random.choice(['SGD(lr='+lr_input + ')',
                             'Adam(lr='+lr_input + ')',
                             'Adagrad(lr='+lr_input + ')',
                             'RMSprop(lr='+lr_input + ')'])

  return bs, init,lr_input, opt
```

## Building & Training the model

```
for _ in range(10):
  print(f"\n************ Details of Model {_+1} ************:")

  # getting the hyper parameters
  param = get_hyperparameter()
  print(param)
  batch_size, init, lr, opt = param[0], param[1], param[2], param[3]

  # creating the MLP model
  model = keras.Sequential(
  [
    keras.Input(shape=(28,28)),   
    layers.Flatten(),
    layers.Dense(64,kernel_initializer=init,activation='relu'),
    layers.Dense(64,activation='relu'),
    layers.Dense(10, activation="softmax"),
  ]
  )
  # model.summary()

  # running the model
  epochs = 10
  model.compile(loss="sparse_categorical_crossentropy", optimizer=eval(opt), metrics=["accuracy"])
  history = model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1)
  score = model.evaluate(x_test, y_test, verbose=2)
  historydf = pd.DataFrame(history.history, index=history.epoch).reset_index()

  # plotting the model info
  print(f"\nModel {_+1} Results:")
  print("Test loss:", score[0])
  print("Test accuracy:", score[1])
  print('\n')
  print('Model Parameters :') 
  print('batch_size={bs} '.format(bs=batch_size))
  print('initializer={i} '.format(i=init))
  print('leraning rate={lr} '.format(lr=lr))
  print('optimizer={opt} '.format(opt=opt))
  print('\n')
  print('Plots:')
  historydf.plot(y=['loss','val_loss'])
  historydf.plot(y=['accuracy','val_accuracy'])
  plt.show()
```
